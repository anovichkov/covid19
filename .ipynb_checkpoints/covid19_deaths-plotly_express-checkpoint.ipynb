# Библиотека plotly express
Использую другую библиотеку для визуализации. Из плюсов: интерактив, подключение системных шрифтов, экспорт растрвых изображений нужного размера. Из минусов: дольше считает.

```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
from datetime import timedelta
import plotly.express as px
import os
from IPython.display import Image
```

## Подключение к данным университета Хопкинса

```python
# ссылку на файл кладу в переменную

url='https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'
```

```python
world = pd.read_csv(url) # читаю csv
world = world.set_index('Country/Region') # делаю название страны индексом
world = world.drop(columns=['Province/State','Lat','Long']) # удаляю колонки с провинцией, широтой и долготой
world.head()
```

```python
world = world.groupby(level=0).sum() # суммирую строки с повторяющимися названиями стран
world = world.sort_values(world.columns[-1], ascending = False) # сортирую по убываню последний столбец
world = world[:5]
world
```

## Перевод

```python
# мне пришлось переводить сочетание "страна США" так как иначе "US" переводилось как "МЫ"


translate = pd.read_csv('translate.csv')

# после "страна " следует удалить

for i in range(len(translate['translate'])):
    translate['translate'][i] = translate['translate'][i][7:]
    
translate = translate.set_index('Country_Region') # делаю столбец 'Country_Region' индексом

translate.head()
```

## Объединение перевода и данных университета

```python
world = pd.concat([world, translate], axis=1, join='inner') # объединение
world = world.set_index('translate') # сделал индексом названия стран на русском
```

```python
world.head()
```

## Транспонирование и вычесление ежедневной смертности

```python
world = world.T # транспонирование таблицы
world = world.diff().stack() # разница между строками и создание плоской таблицы
world = world.reset_index() # сброс индекса
world.head()
```

```python
world = world.rename(columns={"level_0": "Data", "translate": "Country", 0: "Deaths"}) # переименовал столбцы
world.head()
```

## Работа с датами

```python
time = world['Data'] # забрал столбец с датами

world_time=[] # создаю пустой список

for i in time:
   world_time.append(i[:-2] + '20' + i[-2:]) # с помощью цикла дописываю значение года, то есть из 20 делаю 2020. 20 интегрируется таким образом, чтобы в следующем году значение года стало 2021

world_time = pd.to_datetime(world_time, format='%m/%d/%Y') # преобразую в стандартный формат времени

world_time
```

```python
world.index = world_time # заменил данные в индексе на корректные даты
world
```

```python
world = world.drop(columns=['Data']) # удалил столбец с некорректными датами
world
```

```python
now = datetime.datetime.now().strftime('%Y%m%d') # создаю запись даты в нужном формате для имени растрового файла
now
```

```python
world_time[-1] # забираю последнюю корректную дату
```

```python
last_time = world_time[-1].strftime('%d.%m.%Y') # преобразую дату в нужный мне формат для подписи в футере
print(last_time)
```

## Визуализация

```python
# Основные настройки визуализации

#задаем переменную fig к оторой можно применять настройки plotly express и указываем, что у нас будет линия

fig = px.line(
    world, # из какого датафрейма берем данные
    world.index, # что будет на оси Х
    world['Deaths'], # что будет на оси Y
    color=world['Country'], # разбиение по цвету
    
    # нужно задать диапазон оси Х
    
    range_x=[ 
        '2020-02-23', # дата старта
        datetime.datetime.now()], # сегодняшний день
    
    range_y=[
        0, # старт оси Y
        world['Deaths'].max()] # максимальное значение Y равное максимальное значение смертности выбранных стран
              )

# Выбор шаблона и настройка его элементов

fig.update_layout(
    
    template = 'plotly_white', # одна из шаблонных тем plotly
    
    font = {'family': 'Verdana', # меняю основной шрифт
            'color': 'rgb(0,0,0)', # цвет шрифта
            'size': 12 # размер шрифта
           },
    
    width=600, height=600, # размер самой визуализации в пикселях
    
    title={
    'text': "<b>Ежедневная смертность от COVID-19</b>", # заголовок. с помощью html-тега указываем что начертание жирное
    'y':.955, # положение заголовка по Y. величина равна отношению от нижнего края до места заголока к высоте файла
    'x':.03, # положение заголовка по Х. величина равна отношению от левого края до места заголока к ширине файла
    'font':
        {
        'size': 21, # размер заголовка
        'color': 'rgb(0,0,0)' # цвет заголовка
        },
    'xanchor': 'left', # выравнивание заголовка
    },
    
    
    # настройки оси X
    
    xaxis_title="Время", # подпись оси X
    
    xaxis_tickformat = '%d.%m', # формат записи дат на оси
    
    xaxis = dict(
        tickfont=dict(
            color='rgba(100,100,100,1)', # цвет шрифта подписей засечек
            size=10, # размер шрифта подписей засечек
        ),
        ticks="outside", # тип расположения засечек
        tickcolor='rgba(200,200,200,1)', # цвет засечек
        showline=True, # видна ли крайняя линия
        linewidth=1, # толщина крайней линии
        linecolor='rgba(200,200,200,1)', # цвет крайней линии
        showgrid=True, # видны ли линии сетки
        gridcolor='rgba(230,230,230,1)', # цвет линий сетки
        title_standoff = 2, # дистанция между подписями засечек и заголовком оси
    ),
    
    # настройки оси Y
    
    yaxis_title="Умершие за день", # подпись оси Y
    
    yaxis=dict(
        tickfont=dict(
            color='rgba(100,100,100,1)', # цвет шрифта подписей засечек
            size=10, # размер шрифта подписей засечек
        ),
        ticks="outside", # тип расположения засечек
        tickcolor='rgba(200,200,200,1)', # цвет засечек
        showline=True, # видна ли крайняя линия
        linewidth=1, # толщина крайней линии
        linecolor='rgba(200,200,200,1)', # цвет крайней линии
        showgrid=True, # видны ли линии сетки
        gridcolor='rgba(230,230,230,1)', # цвет линий сетки
        title_standoff = 2, # дистанция между подписями засечек и заголовком оси
    ), 
    
    margin=dict(
        b=60,
        l=50,
        t=60,
        r=25
    ),
    
    paper_bgcolor = 'white',
    
    legend=dict(
        x=.05,
        y=.95,
        traceorder="normal",
        font=dict(
            size=12,
            color="black"
        ),
        bgcolor="white",
        bordercolor="rgba(230,230,230,1)",
        borderwidth=1
    ),
    
    annotations=[dict(
            showarrow=False,
            x='2020-02-23',
            y=0,
            text="© dataviz.ru. Источник: Johns Hopkins CSSE. Дата: " + last_time,
            xanchor="left",
            xshift=-65,
            yshift=-70,        
            )]
)

fig.show()
```
